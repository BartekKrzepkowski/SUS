## Walidator

Przyjmując poziom istotnosci rowny $\alpha = 0.9$, wyznaczę oszacowanie błędu generalizacji $BG$ w zależności od wielkości zbioru testowego $|T|$, dlatego, że wielkość zbioru testowego związana jest z $\epsilon > 0$, zależnością $-ln(1-\alpha) = 2 * |T| * \epsilon$.<br>
Wynika z tego, że ustalając $T$, przy wcześniej ustalonymy $\alpha$, wyznaczony zostaje $\epsilon$ oraz błąd empiryczny $BE_{T}$.<br>
Zadanie polega zatem na minimalizacji sumy błędu empirycznego $BE_{T}$ i $\epsilon$<br>
$P(BG < BE_{T} + \epsilon) \geq 1-e^{-2|T|\epsilon}$

```python
def eps_value(alpha, test_size):
    return np.sqrt(-np.log(1 - alpha) / (2 * test_size))
```

```python
def err_plot(clf, alpha, k, to_cycle=5):
    bottom = 1/X.shape[0]
    dependency = []
    for test_freq in tqdm(np.linspace(bottom, 1-bottom, k+2)[1:-1]):
        upper_bound = 0
        for _ in range(to_cycle):
            x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=test_freq)
            sk_nb = clf.fit(x_train, y_train)
            upper_bound += 1-sk_nb.score(x_test, y_test) + eps_value(alpha, y_test.shape[0])
        dependency.append((y_test.shape[0], upper_bound/to_cycle))
        
    dependency = np.array(dependency)
    upped_bound_idx = np.argmin(dependency[:, 1])
    test_size = dependency[upped_bound_idx, 0]
    ge_upper_bound = dependency[upped_bound_idx, 1]
    plt.grid(True)
    plt.xlabel("test_size")
    plt.ylabel("empirical_error + epsilon_value")
    plt.scatter(dependency[upped_bound_idx, 0], dependency[upped_bound_idx, 1], c="k")
    plt.plot(dependency[:,0], dependency[:,1], )
    
    return int(test_size), eps_value(alpha, test_size), ge_upper_bound
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
test_size1, eps_value1, upper_bound1 = err_plot(clf=GaussianNB(), alpha=0.9, k=100)
```

```python
display(Math(r'|T|: {} \\ \epsilon: {:.4f} \\ Błąd\ empiryczny: {:.4f} \\ Najmniejsze\ napotkane\ oszacowanie\ błędu\ generalizacji: {:.4f}'\
             .format(test_size1, eps_value1, upper_bound1-eps_value1, upper_bound1)))
```

Użyte zostały defaultowe parametry:

```python
GaussianNB().get_params()
```

## k-NN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
plt.figure(figsize=(10, 10))
for nbn in range(2, 25, 3):
    test_size2, eps_value2, upper_bound2 = err_plot(clf=KNeighborsClassifier(n_neighbors=nbn), alpha=0.9, k=100)
    display(Math(r'k: {} \\ |T|: {} \\ \epsilon: {:.4f} \\ Błąd\ empiryczny: {:.4f} \\ Najmniejsze\ napotkane\ oszacowanie\ błędu\ generalizacji: {:.4f}'\
             .format(nbn, test_size2, eps_value2, upper_bound2-eps_value2, upper_bound2)))
plt.legend(range(2, 25, 3));
```

Użyte zostały defaultowe parametry, oprócz iterowanej liczby sąsiadów względem których podejmowana jest decyzja:

```python
KNeighborsClassifier().get_params()
```

## Drzewo decyzyjne

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
test_size3, eps_value3, upper_bound3 = err_plot(clf=DecisionTreeClassifier(), alpha=0.9, k=100)
```

```python
display(Math(r'|T|: {} \\ \epsilon: {:.4f} \\ Błąd\ empiryczny: {:.4f} \\ Najmniejsze\ napotkane\ oszacowanie\ błędu\ generalizacji: {:.4f}'\
             .format(test_size3, eps_value3, upper_bound3-eps_value3, upper_bound3)))
```

Użyte zostały defaultowe parametry:

```python
DecisionTreeClassifier().get_params()
```
