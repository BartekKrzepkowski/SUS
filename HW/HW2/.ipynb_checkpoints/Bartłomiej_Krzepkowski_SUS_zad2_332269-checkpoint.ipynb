```python
paths_files = glob.glob(os.path.join(PATH, "*.png"))
paths_files = np.array(paths_files)
```

```python
def display_images(paths_files, nb=100, w=20):
    h = (nb-1)//w + 1
    fig, ax = plt.subplots(h, w, figsize=(10,10))
    paths_files = np.array(paths_files)
    paths_files = np.random.permutation(paths_files)
    for i, path in enumerate(paths_files):
        file = cv2.imread(path)
        ax[i//w][i%w].imshow(file)
        if i == nb-1:
            break
```

```python
display_images(paths_files)
```

```python
from umap import UMAP
from sklearn.cluster import DBSCAN
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
```

```python
def get_representation_keras(paths):
    files = []
    model = VGG16(input_shape=(32,32, 3), include_top=False)
    for path in paths:
        img = image.load_img(path, target_size=(32, 32))
        file = np.expand_dims(image.img_to_array(img), axis=0)
        file = model.predict(file).squeeze()
        files.append(file)
    return np.array(files)
```

```python
files = get_representation_keras(paths_files)
```

```python
stsc = StandardScaler()
files_sc = stsc.fit_transform(files)
files_sc.shape
```

Po wyodrębieniu cech, za pomocą przeuczonej sieci neuronowej, przejdę do klastrowania.

Badam podobieństwo w przestrzeni cech wzgledem dwóch miar - euklidesowej i cosinusowej.<br>
Niech:<br>
> D - wymiarowość danych<br>
> minK - minimalny rozmiar klastra<br>
> $\epsilon$ - promień detekcji członka

W DBSCAN parametry wybiera się:<br>
> minK > D<br>
> $\epsilon$ - punkt kolankowy wykresu k-NN

chce dopuścić klastry liczące niewiele elementów(od 10), dlatego zmniejszam wymiarowość danych za pomocą UMAP

```python
minK = 10
D = minK - 1
```

```python
def min_dist(indices, dist):
    N = dist.shape[0]
    k = (dist[N-1]-dist[0])/N
    f = lambda x: abs(dist[x] - k*x - dist[0])
    g = lambda x: k*x + dist[0]
    plt.figure(figsize=(10, 10))
    plt.scatter(indices, f(indices), label="dist from pend")
    plt.scatter(indices, g(indices), label="pend")
    plt.scatter(indices, dist, label="eps")
    pen = np.array(list(map(f, indices)))
    x_knee = np.argmax(pen)
    y_knee = round(dist[x_knee], 6)
    plt.axvline(x=x_knee, label="{} --x".format(x_knee))
    plt.axhline(y=y_knee, label="{} --y".format(y_knee))
    plt.legend()
    return x_knee, y_knee
```

```python
def find_knee(data, k, metric, frac=1, kick=0):
    OUT = round(kick*data.shape[0]/1000)
    END = round(data.shape[0]*frac)
    nbrs = NearestNeighbors(n_neighbors=k, metric=metric).fit(data)
    distances, _ = nbrs.kneighbors(data)
    distanceDec = np.array(sorted(distances[:, k-1], reverse=True))
    return min_dist(np.arange(END-OUT), distanceDec[OUT:END])
```

# Euclidean metric

```python
metric_e = "euclidean"
files_umap = UMAP(n_components=D, metric=metric_e, target_metric=metric_e).fit_transform(files_sc)
stsc = StandardScaler()
files_umap = stsc.fit_transform(files_umap)
_, eps_e = find_knee(files_umap, minK, metric_e, frac=0.9)
```

```python
#euclidean
db_clust = DBSCAN(eps=eps_e, min_samples=minK, n_jobs=-1, metric=metric_e)
db_clust.fit(files_umap)
df = pd.DataFrame(list(enumerate(db_clust.labels_)), columns=["idx", "labels"]).groupby("labels").count().T
grouped_idx = np.array([el[1].index.values for el in pd.DataFrame(db_clust.labels_, columns=["labels"]).groupby("labels")])
display(HTML(df.to_html()))
```

# Cosine metric

```python
metric_c = "cosine"
files_umap = UMAP(n_components=D, metric=metric_c, target_metric=metric_c).fit_transform(files_sc)
stsc = StandardScaler()
files_umap = stsc.fit_transform(files_umap)
_, eps_c = find_knee(files_umap, minK, metric_c, frac=0.9)
```

```python
db_clust = DBSCAN(eps=eps_c, min_samples=minK, n_jobs=-1, metric=metric_c)
db_clust.fit(files_umap)
df = pd.DataFrame(list(enumerate(db_clust.labels_)), columns=["idx", "labels"]).groupby("labels").count().T
grouped_idx = np.array([el[1].index.values for el in pd.DataFrame(db_clust.labels_, columns=["labels"]).groupby("labels")])
display(HTML(df.to_html()))
```
